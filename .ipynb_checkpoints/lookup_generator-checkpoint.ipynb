```python
import matplotlib.pyplot as plt
import math
```

```python
plt.plot([180, 135, 90, 45, 0], [55000, 100000, 145000, 180000, 185000])
```

```python
elbow_d180 = 55000
elbow_d135 = 100000
elbow_d90 = 145000
elbow_d45 = 180000
elbow_d0 = 185000
```

```python
def elbow_angle_to_bits(x):
    c = (90 - x)/90
    return round(( c * ( elbow_d90 - elbow_d0 ) ) + elbow_d0, 0)
```

```python
def shoulder_angle_to_bits(x):
    c = (90 - x)/90
    return round(( c * ( elbow_d0 - elbow_d90 ) ) + elbow_d90, 0)
```

```python
l = 8  # Inches
```

```python
for x in range(2, 9):
    for y in range(0, 5):
        p = math.tan(y / x) * (180 / math.pi)
        t = math.cos( ( math.sqrt( (x*x) + (y*y) ) / 2 ) / l ) * (180 / math.pi)

        if ( (p+t) > 10 ) and ( (p+t) < 90 ) and ( (p-t) > 10 ) and ( (p-t) < 90 ):
            print(f"x: {x}\ty: {y}\ta1: {p+t}\ta2: {p-t}\ta1: {shoulder_angle_to_bits(p + t)}\ta2: {elbow_angle_to_bits(p - t)}")
```
